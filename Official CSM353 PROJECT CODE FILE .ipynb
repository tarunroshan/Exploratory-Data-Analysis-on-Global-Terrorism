# Global Terrorism Exploratory Data Analysis

## Dataset Description 

This dataset (globalterrorismdb_0718dist.csv) usually contains details on terrorism incidents worldwide and includes columns that cover a range of information such as:

Event Identification: Information about each specific incident, such as event ID, date, and location.

Incident Characteristics: Attributes like attack type, target type, weapons used, and perpetrator information.

Casualties and Damages: Columns that indicate the number of casualties, injuries, and any property damage.

Geopolitical Context: Information on the region, country, and groups involved.

Motives and Outcomes: Ideological or logistical motives behind incidents and the outcomes of the attacks.

This dataset records global terrorism incidents, with each entry uniquely identified by an eventid. It includes the date (iyear, imonth, iday) and approximate date (approxdate) of each incident, as well as location details like country and region. The extended field flags prolonged incidents, and addnotes provides additional context. Source information is given in scite1, scite2, scite3, and dbsource. Motive indicators (INT_LOG, INT_IDEO, INT_MISC, INT_ANY) classify the incident type, while related links to associated events. This structure enables analysis of terrorism trends by time, location, and motive.

### Handling Outliers

The following code Creates a box plot to plot the distribution of nkill (number of kills column). This can identify outliers or extreme values in the data.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='nkill')
plt.title("Box Plot for Number of Kills (nkill)")
plt.xlabel("Number of Kills")
plt.show()

```

This following lines of code removes outliers from DataFrame df; it is using the Interquartile Range (IQR) method and creates a new DataFrame df_no_outliers without outliers.

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
```

```python
df_no_outliers = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
print(df)
print("\nDataFrame after removing outliers:")
print(df_no_outliers)
```

## Feature Engineering



This following code does feature engineering by incorporating new columns into the dataset: 'casualties', 'success_rate', 'attack_class', 'target_class', 'region_group', 'is_suicide', 'high_casualty', 'year_month' to analyze the terrorism-related events more closely.


```python
# Feature Engineering
# 1. Create a 'casualties' feature by adding 'nkill' and 'nwound'
df['casualties'] = df['nkill'].fillna(0) + df['nwound'].fillna(0)
# 2. Create a 'success_rate' feature where 'success' is 1 if the attack was successful, 0 otherwise
df['success_rate'] = df['success'].apply(lambda x: 1 if x == 1 else 0)
# 3. Create an 'attack_class' feature that classifies attack types
attack_mapping = {
    1: 'Assassination', 2: 'Armed Assault', 3: 'Bombing/Explosion',
4: 'Hijacking', 5: 'Hostage Taking', 6: 'Infrastructure Attack',
    7: 'Unarmed Assault', 8: 'Unknown'}
df['attack_class'] = df['attacktype1'].map(attack_mapping)
# 4. Reducing target type categories into more general categories
target_mapping = {
    1: 'Business', 2: 'Government', 3: 'Police', 4: 'Military', 5: 'Abortion Related',
    6: 'Airports & Aircraft', 7: 'Government', 8: 'Religious Figures/Institutions',
    9: 'Terrorists/Militias', 10: 'Telecommunications', 11: 'Tourists', 12: 'Transportation',
13: 'Utilities', 14: 'NGO', 15: 'Journalists', 16: 'Maritime', 17: 'Unknown'
}
df['target_class'] = df['targtype1'].map(target_mapping)
# 5. Creating a 'region_group' feature for region classification based on proximity/geopolitical consideration
region_mapping = {
    1: 'North America', 2: 'Central America', 3: 'South America', 4: 'Western Europe',
    5: 'Eastern Europe', 6: 'Middle East & North Africa', 7: 'Sub-Saharan Africa',
8: 'South Asia', 9: 'Southeast Asia', 10: 'Central Asia', 11: 'East Asia',
12: 'Australia & Oceania', 13: 'Worldwide'
}
df['region_group'] = df['region'].map(region_mapping)
# 6. Binarizing the 'suicide' feature where 1 indicates a suicide attack, 0 otherwise
#
df['is_suicide'] = df['suicide'].apply(lambda x: 1 if x == 1 else 0)
# 7. Generating a 'high_casualty' feature where incidents with casualties > 10 are marked as high casualty events

df['high_casualty'] = df['casualties'].map(lambda x: 1 if x>10 else 0)
# 8. Add feature 'year_month' in order to aggregate by month for better temporal analysis
df['year_month'] = df['iyear'].astype(str)+'-'+df['imonth'].map(lambda x : f'{int(x):02d}')
# Print a few rows to see what the new features look like
df[['casualties','success_rate', 'attack_class', 'target_class', 'region_group', 'is_suicide', 'high_casualty', 'year_month']].head()
```

However, the correlations between casualties - killed and wounded respectively - and success of the attack are very low: 0.05 and 0.023 respectively, suggesting there is little connection between the number of casualties and whether an attack is classified as successful or not.

```python
df_corr = df[['nkill', 'nwound', 'success']].fillna(0).corr()

plt.figure(figsize=(8,6))
sns.heatmap(df_corr, annot=True, cmap='YlGnBu', linewidths=0.5, linecolor='black')
plt.title('Heatmap: Correlation Between Casualties and Attack Success')
plt.show()
```

### Time Series Analysis - Monthly Trend
Analysis: This heatmap depicts the trend of monthly attacks over the period of terrorism attacks from 1970 to 2017. Color intensity is interpreted by number of attacks with darker shades indicating higher counts. A gradual increase in the count of attacks can be seen from mid-2000s that peaked in darker shades around the period 2014-2015. This pattern shows an increase in terrorist activity around these years, possibly due to certain conflicts or geopolitical events.

We further notice that some months tend to have higher counts, especially in the middle of the year, and this might be attributed to some seasons.
```python
df_monthly = df.dropna(subset=['iyear', 'imonth'])
monthly_trend = df_monthly.groupby(['iyear', 'imonth']).size().unstack().fillna(0)

plt.figure(figsize=(14, 7))
sns.heatmap(monthly_trend, cmap='YlOrRd', cbar_kws={'label': 'Number of Attacks'})
plt.title('Monthly Trend in Terrorism Attacks (Heatmap)')
plt.xlabel('Month')
plt.ylabel('Year')
plt.show()
```

### Investigating the Impact of Significant Global Events
Analysis: This stacked bar chart shows, over the two major years, 2001 and 2011, terrorism attacks through months affected by worldwide significant events. In 2001, we can observe a relatively lower number of attacks in general but with seasonal variation throughout the months. This is the year of the 9/11 attacks, which had a huge influence on changing the course of worldwide counter-terrorism strategies but which, as yet, had not contributed to a substantial upsurge in incidences of terrorism across the globe in that year.

In contrast, by 2011 attacks per month are substantially higher in comparison, which would indicate a general uptick in terrorism ten years past 9/11. This is consistent with a variety of geopolitics-driven conflicts and strife throughout much of the Middle East and North Africa-a very tumultuous Arab Spring season. The monthly attack counts for 2011 are in most months merely high, suggesting that at this point terrorism spread much more broadly and no longer was so strongly occasioned to specific seasonal rhythms.

```python
significant_years = [2001, 2011]
impact_trend = df[df['iyear'].isin(significant_years)].groupby(['iyear', 'imonth']).size().unstack()

plt.figure(figsize=(12, 6))
impact_trend.plot(kind='bar', stacked=True, colormap='coolwarm', ax=plt.gca())
plt.title('Impact of Significant Global Events on Terrorism Attacks')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.legend(title='Month')
plt.grid()
plt.show()
```
